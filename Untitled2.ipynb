С помощью MinMaxScalar вычислим минимум и максимум по обучающим данным. Затем создадим новый ДФ основанный на распределении от 0 до 1. Данные для которого мы возьмем из старого набора данных, и изменим их с помощью transform. transform(self, X) скалирует значения X в соответствии с растоянием заначений которые мы получаем от MinMaxScalar.

В scikit-learn есть специальный класс Pipeline, с помощью которого можно создавать конструктор определяющий последовательность из шагов, трансформирующих данные в нужном порядке. Создадим такой-же конвеер как в методичке. Опишем классы для преобразования.

DropOutlet
удаляет выбросы основываясь на правиле 3-х сигм
Опишем методы fit-обучения и transform - изменения данных

```python
import sklearn.pipeline as pipeline
from sklearn.base import BaseEstimator, TransformerMixin

class DropOutlet(BaseEstimator):
    """
    Удаление выбросов, основываясь на правиле 3-х сигм (только для нормального распеделения)
    Основываясь на графиках прошлой лабораторной работы, можно сказать что значения имеют нормальное распространение 
    """
    def fit(self, X: pd.DataFrame, y=None):
        """
        Получаем стандартное отклонение для каждого столбца и сохраняем их        
        """
        df.select_dtypes(include=['float', 'int'])
        self.std = X.std() #получаем стандартное отклонение 
        self.columns = self.std.index.values 
        print( 'Стандартное отклонение :\n',self.std)
        return self
    def transform(self, X: pd.DataFrame, y=None):
        """
        Убираем все здачения, котоые находятся дальше 3-х стандартных отконений (сигма) от мат. ожидания случайной величины
        """
        X_cols = X[self.columns]
        print(X_cols)
        """"
        возвращаем значение - датафрейм без значений котрорые меньше чем разность между значением 
        и мат.ожиданием на 3 стандартных отклонения
        """
        return X[ (X_cols - X_cols.mean()).abs() <= 3*self.std ].dropna()
class PandasToNumpy(BaseEstimator):
    """
    Просто преобразует данные из DataFrame от pandas к матрице от numpy (ndarray)
    """
    
    def fit(self, X : pd.DataFrame = None, y=None):
        return self
    
    def transform(self, X: pd.DataFrame, y=None) -> np.ndarray:
        return X.values
    


```

С помощью команды get_dummies() преобразуем категориальные переменные в индикаторные переменные. Так как таких переменных у нас нет, то команда отработает без преобразований. Создадим конвеер преобразований. Через него проведем копию нашего датафрейма. Конвеер будет состоять из 2 этапов:

    DropOutlet - удаление выбросов в датафрейме.
    
    PandasToNumpy - преобразование датафрейма в матрицу нампай
    
После преобразований выведем резмерность новой таблицы, Количество убранных выбросов. В одном из пунктов конвеера выводим готовую таблицу, чтобы убедиться в правильности измененых данных.

```python
X_data = mm_norm_number_data[mm_norm_number_data.columns].copy()
print(X_data.shape)

X_data = pd.get_dummies(X_data)

pipe_x = pipeline.Pipeline([('drop_outlet', DropOutlet()),('to_matrix', PandasToNumpy())])
number_data_ready_X = pipe_x.fit_transform(X_data)
print('Размерность таблицы',number_data_ready_X.shape, 'Количество убранных выбросов: ',X_data.shape[0]-number_data_ready_X.shape[0])
X_data=number_data_ready_X
```

Как можно заметить, все данные преобразовались в величины от 0 до 1, выбросов оказалось 16 штук. Теперь узнаем что за данные получились в конце вычислений. Как можно заметить, это те же самые данные но в виде массива массивов. Это результат выполнения второго пункта конвеера.

Создадим функцию которую затем задекорируем в контекстмэнеджер. Декоратор начинается с символа @, за которым следует название функции, которую мы собираемся «декорировать» Наша функция будет просто отчитывать время выполнения работы. По окончанию работы функции она выведет время затраченное на работу в секундах.Далее обычим наши данные на алгоритме PCA - склерновский алгоритм анализа основных компонент. согласно оф.сайту склерна алгоритм PCA - Декомпозиция данных для проецирования их в пространство меньших измерений. Таким образом мы получаем данные собранные на 2D пространстве.

```python
from contextlib import contextmanager
from time import time
from sklearn.decomposition import PCA

@contextmanager
def measure_time():
  t = time()
  yield
  print(f'Потрачено времени {time() - t} сек')
print(X_data)
X_ = PCA(n_components=2).fit_transform(X_data)
X0_, X1_ = X_[:, 0], X_[:, 1]
```

На основе первого алгоритма AgglomerativeClustering - Кластеризация методом агломератов. Когда мы создаем один большой кластер и усекаем его на каком-нибудь уровне(Красная черта на дендрограмме) Этот уровень определяет глубину различия кластеров.
Узнаем различные критерии оценки близости: linkage - Какой критерий связи использовать. Критерий связи определяет, какое расстояние использовать между наборами наблюдений:

    ward - (стоит по дефолту, если не указывать другие критерии) уменьшает дисперсию обьединяемых классов.

    average - использует среднее расстояние каждого наблюдения двух классов

    complete - использует максимальные расстояния между всеми наблюдениями двух классов

    single - использует минимум расстояний между всеми наблюдениями двух наборов

![title](img1.png)

```python
from sklearn.cluster import AgglomerativeClustering

AggClu = AgglomerativeClustering(linkage='average')
#создаем новый  
with measure_time():
    AggClu.fit(X_data)

plt.figure(figsize=(10, 10))
plt.scatter(X0_, X1_, c=AggClu.labels_.astype('int'), cmap=plt.cm.coolwarm)
```

Далее алгоритм Сдвига среднего. Он заключается в 5 шагах :
    
    Шаг 1 – Сначала начните с точек данных, назначенных их кластеру.

    Шаг 2 – Далее этот алгоритм будет вычислять центроиды.

    Шаг 3 – На этом шаге местоположение новых центроидов будет обновлено.

    Шаг 4 – Теперь процесс будет повторен и перемещен в область более высокой плотности.

    Шаг 5 – Наконец, он будет остановлен, как только центроиды достигнут позиции, из которой он не сможет двигаться дальше.

```python
from sklearn.cluster import MeanShift

MS = MeanShift()
with measure_time():
    MS.fit(X_data)
    cluster_centers = MS.cluster_centers_

plt.figure(figsize=(10, 10))
plt.scatter(X0_, X1_, c=MS.labels_.astype('int'), cmap=plt.cm.coolwarm)
plt.scatter(cluster_centers[:,0],cluster_centers[:,1], c='k', linewidths = 7)
print(MS.labels_.mean())       
    
```

Сдвиг среднего не отработал как нужно, не определив какое-либо количество кластеров. В отличие от кластеризации K-средних, она не делает никаких предположений; следовательно, это непараметрический алгоритм. Это значит что нельзя задать количество кластеров напрямую. Увы.
